```python
import pandas as pd
import re
import matplotlib.pyplot as plt
plt.style.use("bmh")
plt.rcParams["figure.figsize"] = (8, 6)
plt.rcParams["figure.dpi"]     = 150
```

```python
def chamfer_distance_mean(df):
    return (df["chamfer_distance0"].mean() + df["chamfer_distance1"].mean()) / 2

def dice2d_mean(df):
    return (df["dice2d0"].mean() + df["dice2d1"].mean()) / 2

def dice2d_backproj(df):
    return (df["bp_dice2d0"].mean() + df["bp_dice2d1"].mean()) / 2,

def chamfer_distance_mm_mean(df):
    return (df["chamfer_distance0_mm"].mean() + df["chamfer_distance1_mm"].mean()) / 2

def chamfer_distance_backproj(df):
    return (
        (df["bp_chamfer_distance0"].mean() + df["bp_chamfer_distance1"].mean()) / 2,
        (df["bp_chamfer_distance0_mm"].mean() + df["bp_chamfer_distance1_mm"].mean()) / 2,
    )
```

```python
df = pd.read_csv("../data/clinical_right.csv")
df_baseline = pd.read_csv("../data/clinical_baseline_right.csv")
```

```python
print(
    "CD[px]=", chamfer_distance_mean(df), "CD[mm]=", chamfer_distance_mm_mean(df),
    "CD_b[px]=", chamfer_distance_mean(df_baseline), "CD_b[mm]=", chamfer_distance_mm_mean(df_baseline)
)
```
