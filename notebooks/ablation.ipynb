```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
import numpy as np

plt.style.use("bmh")
plt.rcParams["figure.figsize"] = (8, 6)
plt.rcParams["figure.dpi"]     = 150
```

```python

def mean_ci(series, n_boot=10000, confidence=0.95, seed=42):
    rng = np.random.default_rng(seed)
    data = series.dropna().values
    means = [np.mean(rng.choice(data, size=len(data), replace=True)) for _ in range(n_boot)]
    lower, upper = np.percentile(means, [(1 - confidence) / 2 * 100, (1 + confidence) / 2 * 100])
    return np.mean(data), lower, upper

def ci_pretty(df, col):
    mean, lo, hi = mean_ci(df[col])
    print(f"{mean:.2f} ({lo:.2f}-{hi:.2f})")
```

# Guidance scale

```python
def plot_w_guidance(gds, dfs, label, ci=0.95):
    means, cis = [], []

    for df in dfs:
        data = df["refined_chamfer_adaptive"].dropna()
        mean = data.mean()
        sem = stats.sem(data)
        h = sem * stats.t.ppf((1 + ci) / 2., len(data) - 1)
        means.append(mean)
        cis.append(h)

    means = np.array(means)
    cis = np.array(cis)
    plt.errorbar(gds, means, yerr=cis, fmt='o', capsize=4, label=label, alpha=0.5)
```

```python
df_guidance0 = pd.read_csv("../data/synthetic_guidance_scale_0.0_right.csv")
df_guidance1 = pd.read_csv("../data/synthetic_guidance_scale_1.0_right.csv")
df_guidance2 = pd.read_csv("../data/synthetic_guidance_scale_2.0_right.csv")
df_guidance6 = pd.read_csv("../data/synthetic_guidance_scale_6.0_right.csv")
df_guidance8 = pd.read_csv("../data/synthetic_guidance_scale_8.0_right.csv")
df_guidance10 = pd.read_csv("../data/synthetic_guidance_scale_10.0_right.csv")
df_guidance12 = pd.read_csv("../data/synthetic_guidance_scale_12.0_right.csv")

plot_w_guidance(
    [0, 1, 2, 6, 8, 10, 12],
    [df_guidance0, df_guidance1, df_guidance2, df_guidance6, df_guidance8, df_guidance10, df_guidance12],
    label="Right side"
)

df_guidance0 = pd.read_csv("../data/synthetic_guidance_scale_0.0_left.csv")
df_guidance1 = pd.read_csv("../data/synthetic_guidance_scale_1.0_left.csv")
df_guidance2 = pd.read_csv("../data/synthetic_guidance_scale_2.0_left.csv")
df_guidance6 = pd.read_csv("../data/synthetic_guidance_scale_6.0_left.csv")
df_guidance8 = pd.read_csv("../data/synthetic_guidance_scale_8.0_left.csv")
df_guidance10 = pd.read_csv("../data/synthetic_guidance_scale_10.0_left.csv")
df_guidance12 = pd.read_csv("../data/synthetic_guidance_scale_12.0_left.csv")

plot_w_guidance(
    [0, 1, 2, 6, 8, 10, 12],
    [df_guidance0, df_guidance1, df_guidance2, df_guidance6, df_guidance8, df_guidance10, df_guidance12],
    label="Left side"
)

plt.xlabel("guidance")
plt.ylabel("Chamfer distance (95% CI)")
plt.legend()
plt.show()
```
