```python
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use("bmh")
plt.rcParams["figure.figsize"] = (8, 6)
plt.rcParams["figure.dpi"]     = 150
```

# Guidance scale

```python
def plot_w_guidance(gds, dfs):
    means = []
    for df in dfs:
        mean = df["refined_chamfer_adaptive"].mean()
        means.append(mean)
    plt.scatter(gds, means)
```

```python
df_guidance0 = pd.read_csv("../data/guidance_scale_0.0.csv")
df_guidance1 = pd.read_csv("../data/guidance_scale_1.0.csv")
df_guidance2 = pd.read_csv("../data/guidance_scale_2.0.csv")
df_guidance10 = pd.read_csv("../data/synthetic_adaptive_right.csv")
```

```python
plot_w_guidance(
    [0, 1, 2, 10],
    [df_guidance0, df_guidance1, df_guidance2, df_guidance10]
)
```
