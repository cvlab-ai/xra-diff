```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
import numpy as np

plt.style.use("bmh")
plt.rcParams["figure.figsize"] = (8, 6)
plt.rcParams["figure.dpi"]     = 150
```

```python

def mean_ci(series, n_boot=10000, confidence=0.95, seed=42):
    rng = np.random.default_rng(seed)
    data = series.dropna().values
    means = [np.mean(rng.choice(data, size=len(data), replace=True)) for _ in range(n_boot)]
    lower, upper = np.percentile(means, [(1 - confidence) / 2 * 100, (1 + confidence) / 2 * 100])
    return np.mean(data), lower, upper

def ci_pretty(df, col):
    mean, lo, hi = mean_ci(df[col])
    print(f"{mean:.2f} ({lo:.2f}-{hi:.2f})")
```

# Guidance scale

```python
def plot_w_guidance(gds, dfs, label, ci=0.95, column="refined_chamfer_adaptive"):
    means, cis = [], []

    for df in dfs:
        data = df[column].dropna()
        mean = data.mean()
        sem = stats.sem(data)
        h = sem * stats.t.ppf((1 + ci) / 2., len(data) - 1)
        means.append(mean)
        cis.append(h)

    means = np.array(means)
    cis = np.array(cis)
    plt.errorbar(gds, means, yerr=cis, fmt='o', capsize=4, label=label, alpha=0.5)
```

```python
right_guidance0 = pd.read_csv("../data/synthetic_guidance_scale_0.0_right.csv")
right_guidance0_2 = pd.read_csv("../data/synthetic_guidance_scale_0.2_right.csv")
right_guidance0_3 = pd.read_csv("../data/synthetic_guidance_scale_0.3_right.csv")
right_guidance0_5 = pd.read_csv("../data/synthetic_guidance_scale_0.5_right.csv")

plot_w_guidance(
    [0, 0.2, 0.3, 0.5],
    [right_guidance0, right_guidance0_2, right_guidance0_3, right_guidance0_5],
    label="right side",
    column="refined_Ot(2)_adaptive"
)

left_guidance0 = pd.read_csv("../data/synthetic_guidance_scale_0.0_left.csv")
left_guidance0_2 = pd.read_csv("../data/synthetic_guidance_scale_0.2_left.csv")
left_guidance0_3 = pd.read_csv("../data/synthetic_guidance_scale_0.3_left.csv")
left_guidance0_5 = pd.read_csv("../data/synthetic_guidance_scale_0.5_left.csv")

plot_w_guidance(
    [0, 0.2, 0.3, 0.5],
    [left_guidance0, left_guidance0_2, left_guidance0_3, left_guidance0_5],
    label="left side",
    column="refined_Ot(2)_adaptive"
)

plt.title("[Synthetic, 3D]")
plt.xlabel("guidance")
plt.ylabel("O(2) (95 % CI)")
plt.legend()
plt.show()
```
