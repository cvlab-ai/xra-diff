```python

def mean_ci(series, n_boot=10000, confidence=0.95, seed=42):
    rng = np.random.default_rng(seed)
    data = series.dropna().values
    means = [np.mean(rng.choice(data, size=len(data), replace=True)) for _ in range(n_boot)]
    lower, upper = np.percentile(means, [(1 - confidence) / 2 * 100, (1 + confidence) / 2 * 100])
    return np.mean(data), lower, upper

def ci_pretty(df, col):
    mean, lo, hi = mean_ci(df[col].replace(1e7, np.nan))
    print(f"{mean:.2f} ({lo:.2f}-{hi:.2f})")
```

# Guidance scale

```python
def plot_w_guidance(gds, dfs, label, ci=0.95, column="refined_chamfer_adaptive"):
    means, cis = [], []

    for df in dfs:
        data = df[column].replace(1e7, np.nan).dropna()
        mean = data.mean()
        sem = stats.sem(data)
        h = sem * stats.t.ppf((1 + ci) / 2., len(data) - 1)
        means.append(mean)
        cis.append(h)

    means = np.array(means)
    cis = np.array(cis)
    plt.errorbar(gds, means, yerr=cis, fmt='o', capsize=4, label=label, alpha=0.5)
```

# Sampling steps

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats

def boxplot(gds, dfs, label, column="refined_chamfer_adaptive", c="red", shift=0, width=0.35):
    """
    Draws a group of boxplots for each guidance value, allowing multiple groups
    (e.g., 'Left' and 'Right') to be shown side by side.

    Args:
        gds: list of guidance values (x-axis positions)
        dfs: list of pandas DataFrames (one per guidance value)
        label: legend label for this group
        column: metric column to plot
        c: color for the boxes
        shift: horizontal offset (negative/positive) for side-by-side grouping
        width: box width
    """
    data = []
    for df in dfs:
        d = df[column].replace(1e7, np.nan).dropna()
        data.append(d.values)

    # compute shifted positions
    positions = np.arange(len(gds)) + shift

    # make the boxplots
    bp = plt.boxplot(
        data,
        positions=positions,
        widths=width,
        patch_artist=True,
        notch=True,
        boxprops=dict(facecolor=c, color=c, alpha=0.5),
        capprops=dict(color=c, alpha=0.5),
        whiskerprops=dict(color=c, alpha=0.5),
        flierprops=dict(color=c, markeredgecolor=c, alpha=0.5),
        medianprops=dict(color='black', linewidth=1.2),
    )

    # add a fake line to show in legend
    plt.plot([], [], color=c, label=label, linewidth=6, alpha=0.5)

    plt.xticks(np.arange(len(gds)), gds)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

def line():
    steps = np.array([1, 5, 10, 50, 100])

    means_r, cis_r = [], []
    for df in [right_n1, right_n5, right_n10, right_n50, right_n100]:
        data = df["elapsed"].replace(1e7, np.nan).dropna()
        means_r.append(data.mean())
        sem = stats.sem(data)
        h = sem * stats.t.ppf((1 + 0.95) / 2., len(data) - 1)
        cis_r.append(h)

    means_l, cis_l = [], []
    for df in [left_n1, left_n5, left_n10, left_n50, left_n100]:
        data = df["elapsed"].replace(1e7, np.nan).dropna()
        means_l.append(data.mean())
        sem = stats.sem(data)
        h = sem * stats.t.ppf((1 + 0.95) / 2., len(data) - 1)
        cis_l.append(h)

    means_r = np.array(means_r)
    means_l = np.array(means_l)

    slope_r, intercept_r, r_value_r, p_value_r, std_err_r = stats.linregress(steps, means_r)
    slope_l, intercept_l, r_value_l, p_value_l, std_err_l = stats.linregress(steps, means_l)

    plt.figure(figsize=(7,5))
    plt.plot(steps, means_r, 'o-', label=f"Right side (slope={slope_r:.2f})", alpha=0.5)
    plt.fill_between(steps, means_r - np.array(cis_r), means_r + np.array(cis_r), alpha=0.1)

    plt.plot(steps, means_l, 's--', label=f"Left side (slope={slope_l:.2f})", alpha=0.5)
    plt.fill_between(steps, means_l - np.array(cis_l), means_l + np.array(cis_l), alpha=0.1)

    plt.title("[Synthetic, 3D]")
    plt.xlabel("Sampling Steps")
    plt.ylabel("Runtime [s]")
    plt.grid(True, linestyle='--', alpha=0.4)
    plt.legend()
    plt.tight_layout()
    plt.show()

    print(f"Right side slope: {slope_r:.4f}, R={r_value_r**2:.3f}")
    print(f"Left  side slope: {slope_l:.4f}, R={r_value_l**2:.3f}")
line()
```
